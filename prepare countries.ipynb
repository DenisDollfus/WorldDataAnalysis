```python
import pandas as pd
import json
```

Read countries

```python
with open('data/country_list_2022.json','r') as f:
    data = json.loads(f.read())
    
countries = pd.json_normalize(data, record_path =['data'])
countries = countries.rename(columns={"Code": "Code2"})

display(countries.head(3))
display(countries.describe())

print('--- WORLD BANK COUNTRIES ---')
# Countries with 3 letter code and region, from worldbank https://data.worldbank.org/indicator/SP.DYN.LE00.IN 
countries_wb = pd.read_csv('data/life_expectancy/Metadata_Country_API_SP.DYN.LE00.IN_DS2_en_csv_v2_3930088.csv', skiprows=[i for i in range(1,4)], index_col=False)
countries_wb = countries_wb.drop(columns=['Unnamed: 5'])
countries_wb = countries_wb.rename(columns={'Country Code': 'Code3', 'TableName':'Name'})
display(countries_wb.head(3))
display(countries_wb.describe())

print(' ---- MERGED ----')
merged = pd.merge(countries, countries_wb, on='Name')
merged = merged.drop(columns=['SpecialNotes'])
display(merged.head(3))

# finally, countries base is saved:
countries = merged
countries.to_csv('data/processed/countries.csv', index = False)

```
